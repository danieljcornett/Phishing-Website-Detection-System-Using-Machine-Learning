```python
from nltk.tokenize import RegexpTokenizer
```

```python
tokenizer = RegexpTokenizer(r'[A-Za-z]+')
```

```python
#0 element of the URL section of db
df.URL[0]
```

```python
tokenizer.tokenize(df.URL[0])
```

```python
# Tokenize all the URLs
df['text_tokenized'] = df.URL.map(lambda t: tokenizer.tokenize(t))
```

```python
df.head()
```

```python
from nltk.stem.snowball import SnowballStemmer
```

```python
stemmer = SnowballStemmer('english')
```

```python
# Take each tokenized into stems
df['text_stemmed'] = df['text_tokenized'].map(lambda l: [stemmer.stem(word) for word in l] )
```

```python
df.head()
```

```python
# Create new column
df['text'] = df['text_stemmed'].map(lambda l: ' '.join(l))
```

```python
df.head()
```

```python
# Differentiate between good/bad sites
good_sites = df[df.Label == 'good']
bad_sites = df[df.Label == 'bad']
```

```python

from sklearn.feature_extraction.text import CountVectorizer
```

```python
# Convert to vectors
cv = CountVectorizer()
```

```python
# Train
features = cv.fit_transform(df.text)
```

```python
#Slice to show starting 5 rows
features[:5].toarray()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test = train_test_split(features, df.Label, test_size=0.2)
```

## Model Training

```python
from sklearn.linear_model import LogisticRegression
```

```python
l_model = LogisticRegression()
```

```python
l_model.fit(x_train,y_train)
```

```python
# Accuracy of Logistica Regression model
l_model.score(x_test,y_test)
```

```python
# Training accuracy on model
l_model.score(x_train, y_train)
```
